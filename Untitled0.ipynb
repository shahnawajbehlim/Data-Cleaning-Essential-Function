# Data Cleaning & Essential Functions

**Question 1. What is data cleaning, and why is it important in data analysis?**

* What are the potential consequences of analyzing unclean or messy
* Explain the common steps involved in cleaning and organizing data.

Answer - Data cleaning is the process of fixing or removing incorrect, incomplete, duplicate, or irrelevant data from a dataset.
It's important to make data easier to analyze and visualize & to help in making reliable decisions.

2. Consequences of Using Unclean (Messy) Data
- incomplete analysis
- confusion (e.g., dates in different formats)
- business decisions based on bad data
- wrong insights or trends

3. Common Steps in Data Cleaning & Organizing

Here are the standard steps:

Remove Duplicate Data, Handle Missing Values, Fix Errors, Standardize Formats, Remove Irrelevant Data, Delete unnecessary columns or rows, Validate Data, Organize Data

**Question 3. Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.**

Answer - 1. TRIM Function : Removes extra spaces from text (except single spaces between words)
2. LEFT Function : Extracts a specific number of characters from the beginning (left side) of text.
3. RIGHT Function : Extracts characters from the end (right side) of text.
4. MID Function : Extracts text from the middle, starting at a specific position.
5. CONCAT (or CONCATENATE) Function : Combines text from multiple cells into one.

**Question 4. What is the role of date functions like TODAY in managing datasets?**

Answer - TODAY is used to automatically return the current date, helping track timelines, calculate durations, and keep datasets dynamically updated.

**Question 8. Understanding TODAY() and NOW()**

**a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use each function.**

Answer : TODAY() - Returns current date only (When tracking deadlines or days remaining)
NOW() - Returns current date + time (When logging timestamps (report generated time)

**If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are  left until that date?**

Answer : If A1 = 2025-06-10:

=A1 - TODAY()

**c. Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY HH:MM AM/PM.**

Answer : =TEXT(NOW(), "MM/DD/YYYY HH:MM AM/PM")

**d. If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain.**

Answer: If a cell contains =TODAY(), it will automatically update to the new current date when the worksheet is reopened.
